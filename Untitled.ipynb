```python
import pandas as pd
import numpy as np

# ---------------------------
# LOAD DATA
# ---------------------------
df = pd.read_csv("taxi_dataset.csv")

print("Columns in dataset:\n", df.columns)

# ---------------------------
# FIND DATETIME COLUMN AUTOMATICALLY
# ---------------------------
datetime_col = None

for col in df.columns:
    if "date" in col.lower() or "time" in col.lower():
        datetime_col = col
        break

if datetime_col is not None:
    print("\nUsing datetime column:", datetime_col)

    df[datetime_col] = pd.to_datetime(df[datetime_col], errors='coerce')

    # Extract date & hour
    df['date'] = df[datetime_col].dt.date
    df['hour'] = df[datetime_col].dt.hour

    # Grouping
    daily_trips = df.groupby('date').size()
    hourly_trips = df.groupby('hour').size()

    print("\nTrips per Day:\n", daily_trips.head())
    print("\nTrips per Hour:\n", hourly_trips.head())

else:
    print("\n❌ No datetime column found")

# ---------------------------
# BASIC CLEANING
# ---------------------------
df = df.dropna()

if 'trip_distance' in df.columns:
    df = df[df['trip_distance'] > 0]

if 'fare_amount' in df.columns:
    df = df[df['fare_amount'] > 0]

print("\nCleaned Data Shape:", df.shape)

# ---------------------------
# ANALYSIS
# ---------------------------
if 'fare_amount' in df.columns:
    print("\nAverage Fare:", df['fare_amount'].mean())

if 'trip_distance' in df.columns and 'fare_amount' in df.columns:
    print("Correlation:", df['trip_distance'].corr(df['fare_amount']))

# ---------------------------
# ACT 1 + SIMULATION
# ---------------------------
w, x, y, z = 0, 3, 1, 4

def digital_root(n):
    while n > 9:
        n = sum(map(int, str(n)))
    return n

R = digital_root(w + x + y + z)

if 'fare_amount' in df.columns:
    df['new_fare'] = df['fare_amount'] * (1 + R/100)

    print("\nOld Avg Fare:", df['fare_amount'].mean())
    print("New Avg Fare:", df['new_fare'].mean())

# ---------------------------
# SAVE OUTPUT
# ---------------------------
df.to_csv("final_output.csv", index=False)

print("\n✅ CODE RUN SUCCESSFULLY (NO ERRORS)")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------
# LOAD DATA
# ---------------------------
df = pd.read_csv("taxi_dataset.csv")

# ---------------------------
# CLEAN DATA
# ---------------------------
df = df.dropna()

# ---------------------------
# FIND DATETIME COLUMN
# ---------------------------
datetime_col = None
for col in df.columns:
    if "date" in col.lower() or "time" in col.lower():
        datetime_col = col
        break

if datetime_col is None:
    print("❌ No datetime column found")
    exit()

print("Using datetime column:", datetime_col)

# ---------------------------
# CONVERT TO DATETIME
# ---------------------------
df[datetime_col] = pd.to_datetime(df[datetime_col], errors='coerce')
df = df.dropna(subset=[datetime_col])

# ---------------------------
# EXTRACT FEATURES
# ---------------------------
df['hour'] = df[datetime_col].dt.hour
df['date'] = df[datetime_col].dt.date

# ---------------------------
# CORRECT COUNTING (FIXED)
# ---------------------------
hourly_trips = df['hour'].value_counts().sort_index()
daily_trips = df['date'].value_counts().sort_index()

print("\nTrips per Hour:\n", hourly_trips)
print("\nTrips per Day:\n", daily_trips.head())

# ---------------------------
# PLOTTING
# ---------------------------
hourly_trips.plot()
plt.title("Trips per Hour")
plt.xlabel("Hour")
plt.ylabel("Trips")
plt.show()

daily_trips.plot()
plt.title("Trips per Day")
plt.xlabel("Date")
plt.ylabel("Trips")
plt.show()

# ---------------------------
# PEAK HOUR
# ---------------------------
print("\nPeak Hour:", hourly_trips.idxmax())

print("\n✅ ACT 4 DONE")
```

```python
import matplotlib.pyplot as plt

# --- convert datetime safely ---
df[datetime_col] = pd.to_datetime(df[datetime_col], errors='coerce')
df = df.dropna(subset=[datetime_col])

# --- extract features ---
df['hour'] = df[datetime_col].dt.hour
df['date'] = df[datetime_col].dt.floor('D')   # <-- FIX for daily

# --- counts ---
hourly_trips = df['hour'].value_counts().sort_index()
daily_trips = df['date'].value_counts().sort_index()

print("Rows after clean:", len(df))
print("Unique dates:", df['date'].nunique())

# --- PLOTS ---
plt.figure()
hourly_trips.plot()
plt.title("Trips per Hour")
plt.xlabel("Hour")
plt.ylabel("Trips")
plt.show()

plt.figure()
daily_trips.plot()
plt.title("Trips per Day")
plt.xlabel("Date")
plt.ylabel("Trips")
plt.show()

# --- peak hour ---
print("Peak Hour:", hourly_trips.idxmax())
```
